# Preprocessing

## Deskewing

```python
def moments(image):
    c0,c1 = np.mgrid[:image.shape[0],:image.shape[1]] # A trick in numPy to create a mesh grid
    totalImage = np.sum(image) #sum of pixels
    m0 = np.sum(c0*image)/totalImage #mu_x
    m1 = np.sum(c1*image)/totalImage #mu_y
    m00 = np.sum((c0-m0)**2*image)/totalImage #var(x)
    m11 = np.sum((c1-m1)**2*image)/totalImage #var(y)
    m01 = np.sum((c0-m0)*(c1-m1)*image)/totalImage #covariance(x,y)
    mu_vector = np.array([m0,m1]) # Notice that these are \mu_x, \mu_y respectively
    covariance_matrix = np.array([[m00,m01],[m01,m11]]) # Do you see a similarity between the covariance matrix
    return mu_vector, covariance_matrix
```

```python
def deskew(image):
    c,v = moments(image)
    alpha = v[0,1]/v[0,0]
    affine = np.array([[1,0],[alpha,1]])
    ocenter = np.array(image.shape)/2.0
    offset = c-np.dot(affine,ocenter)
    return interpolation.affine_transform(image,affine,offset=offset)
```

```python
digitsDeskew = digitPics
```

```python
for numbers in range (digitsDeskew.shape[0]):
    digitsDeskew.iloc[numbers] = deskew(np.asarray(digitPics.iloc[numbers]).reshape(16,15)).reshape(1,240)
```

## PCA

```python
postPCAPics = digitPicsNorm
```

```python
pca = PCA(random_state=0)
pca.fit(postPCAPics)
fig, ax = plt.subplots(figsize=(15, 15))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.grid(b=True, which='both', axis='both')
plt.show()
```

```python
#extract top 100 components (approx. 95% variance)
topFeatures = pca.transform(postPCAPics)[:, :100]
```

```python
#create features' names
PCANumbers = np.arange(1, 101)
PCAFeatures = list(str(' ') * 100)

for i in range (len(PCANumbers)):
    PCAFeatures[i] = "pc" + str(PCANumbers[i])
    
postPCAPics = pd.DataFrame(topFeatures, columns = PCAFeatures)
```

```python
acc = []

for i in range (100):
    #split the data into train and test
    X = postPCAPics # Features
    y = digitLabels # Target variable
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
    # Create Decision Tree classifer object
    DT = DecisionTreeClassifier(max_depth = 5) #setting max_depth to 5 didnt lead to a salient change in accuracy (-2.5)
    # Train Decision Tree Classifer
    DT = DT.fit(X_train, y_train)
    #Predict the response for test dataset
    y_pred = DT.predict(X_test)
    #check accuracy
    acc.append(metrics.accuracy_score(y_test, y_pred))
    
mean(acc)
```

## Adding random noise

Since the data is normalized, we'll add random Gaussian noise with mean = 0.2 and Sd = 0.05

```python
#load the data in - should be 2000*240
postPCAPics_noise = postPCAPics
```

```python
for PCs in range (postPCAPics_noise.shape[1]):
    m = mean(postPCAPics_noise.iloc[:,PCs])
    s = (abs(postPCAPics_noise.iloc[:,PCs].min()) + abs(postPCAPics_noise.iloc[:,PCs].max())) / 14
    
    for numbers in range (postPCAPics_noise.shape[0]):
        noise = np.random.normal(m,s)
        postPCAPics_noise.iloc[numbers, PCs] = postPCAPics_noise.iloc[numbers, PCs] + noise
```

### Check - did noise help?

```python
acc = []

for i in range (100):
    #split the data into train and test
    X = postPCAPics_noise # Features
    y = digitLabels # Target variable
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
    # Create Decision Tree classifer object
    DT = DecisionTreeClassifier(max_depth = 5) #setting max_depth to 5 didnt lead to a salient change in accuracy (-2.5)
    # Train Decision Tree Classifer
    DT = DT.fit(X_train, y_train)
    #Predict the response for test dataset
    y_pred = DT.predict(X_test)
    #check accuracy
    acc.append(metrics.accuracy_score(y_test, y_pred))
    
mean(acc)
```
